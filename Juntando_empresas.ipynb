# ** Pontos a tratar**

Antes de começar faremos um coluna agregada a cada um de nossos dataframe para que eles possam ser identidicado no futuro os lucros e productos mais vendidos por cada empresa.

```python
df_ea['Empresa'] = 'Empresa A'

df_ea
```

```python

#Este passo foi agregado posteriormente pois não estava tendo uma leitura adecuada a data quando estava no power bi
df_eb['Empresa'] = 'Empresa B'
df_eb['Data da Venda'] = df_eb['Data da Venda'].apply(lambda x: x.split(' ')[0])

```

# **Resposta ponto 2**
Considerando que tem as mesmas colunas com tipos de dados para ser modificados começaremos pela tratativa da tabela da empresa A

--foi falado os responsaveis da empresa A e relataran que todas as vendas feitas de da categoria roupas foram feitas no ano 2022 o resto foi vendido no ano 2023

```python
#aqui observamos as categorias que tem de vendas
df_ea['Categoria do Produto'].unique()
#observamos que foram agregados 3 novas categorias no ano 2023
```

```python
#com este codigo vamos analisar quais são as linhas do dataframe que tem as categorias de roupas vendidas no 2022
inx = df_ea[df_ea['Categoria do Produto'] == 'Roupas'].index
inx
```

```python
df_ea.loc[inx, "Ano_de_venda"] = '2022-01-01'
df_ea
```

```python
#neste processo repetimos
idxe = df_ea[df_ea['Categoria do Produto'] == 'Eletrodomésticos'].index
idxt = df_ea[df_ea['Categoria do Produto'] == 'Eletrônicos'].index
idxa = df_ea[df_ea['Categoria do Produto'] == 'Alimentos'].index
```

```python
#anexamos todos os dados para o valor de ano
df_ea.loc[idxe, "Ano_de_venda"] = '2023-01-01'
df_ea.loc[idxt, "Ano_de_venda"] = '2023-01-01'
df_ea.loc[idxa, "Ano_de_venda"] = '2023-01-01'
df_ea[['Ano_de_venda']]

#com esta etapa acabamos o processo de integrar a etapa de data a nossa empresa A
```

# **Solução do ponto 1**

Neste ponto vamos fazer que os as os dois dataframe tenham os mesmos nomes de columna

```python
c_1 = list(df_eb.keys())
c_2 = list(df_ea.keys())
columns = dict(zip(c_1,c_2))
columns
```

```python
df_eb_new = df_eb.rename(columns=columns)
#empresa_ = dict(empresa_b)
#empresa_

##empresa_b.info()
#df.info()
df_eb_new
```

# **Ponto 3 resolução**

neste ponto vamos ter que juntar as empresas para ter um so dataframe que possar ser enviado para o departamento de analistas de dados

```python
df_empresa_nova = df_ea.append(df_eb_new, ignore_index=True)
df_empresa_nova
```

# **Solução do ponto 4**

para o ponto 4 foi falado que precisa ser feita uma columa de valor total em vendas

```python
df_empresa_nova['valor_Total'] = df_empresa_nova['Preço do Produto (R$)'].multiply(df_empresa_nova['Quantidade em Estoque'].astype(float))
df_empresa_nova['valor_Total'] = df_empresa_nova['valor_Total'].round(2)
df_empresa_nova['valor_Total'][7]
```

# **Solução Ponto 5**

neste ponto vamos criar o arquivo csv para levar ao departamento de analista e eles criar um bi

```python
df_empresa_nova.to_csv('nova_empresa.csv')
df_empresa_nova.to_excel('nova_empresa_excel.xlsx')
```
